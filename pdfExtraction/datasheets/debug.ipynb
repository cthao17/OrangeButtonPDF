```python
# I have JSON files in the OB format, and I want to convert them to the JSON to pass to my frontend when a certain endpoint is called. 
# if possible i want to pass multiple dictionaries/JSON objects to the frontend in a JSON array.
# sample files are in ./train_test/Sheet1/output/______.json (all the json files in this output directory are what i want to put in the JSON array)
import os, json
filepaths = []
def getJSONFilePaths():
    paths = []
    for file in os.listdir("./train_test/Sheet1/output"):
        if file.endswith(".json"):
            paths.append(f"./train_test/Sheet1/output/{file}")
    return paths

filepaths = getJSONFilePaths()
print(filepaths)
```

```python
def parse_OB_JSON(file):
    with open(file, 'r') as f:
        data = json.load(f)
    return data
```

```python
d = parse_OB_JSON(filepaths[0])
print("fields in dictionary: ", len(d["ProdModule"]))
# this step removes the "Value" key from the dictionary if it is empty or -1
deleted = []
kept = []
for key in d["ProdModule"]:
    if type(d["ProdModule"][key]) == dict:
        if "Value" in d["ProdModule"][key].keys() and (d["ProdModule"][key]["Value"] == "" or d["ProdModule"][key]["Value"] == -1):
            deleted.append(d["ProdModule"][key])
            del d["ProdModule"][key]["Value"]
        else:
            kept.append(d["ProdModule"][key])
print("deleted: ", len(deleted))
print("kept: ", len(kept))
print("fields in dictionary: ", len(d["ProdModule"]))

# previous step leaves us with key: {<empty dictionary>}, we want to remove those k:v pairs
for key in d["ProdModule"].copy():
    if d["ProdModule"][key] == {}:
        del d["ProdModule"][key]
print("fields in dictionary: ", len(d["ProdModule"]))
```

```python
data = []
for file in filepaths:
    with open(file, "r") as f:
        d = json.load(f)
        # removes the "Value" key from the dictionary if it is empty or -1
        for key in d["ProdModule"]:
            if type(d["ProdModule"][key]) == dict:
                if "Value" in d["ProdModule"][key].keys() and (d["ProdModule"][key]["Value"] == "" or d["ProdModule"][key]["Value"] == -1):
                    del d["ProdModule"][key]["Value"]
        # previous step leaves us with key: {<empty dictionary>}, we want to remove those k:v pairs
        for key in d["ProdModule"].copy():
            if d["ProdModule"][key] == {}:
                del d["ProdModule"][key]
        data.append(d)
result = json.dumps(data, indent=2)
print(result)
```
